```python
import pandas as pd
import numpy as np
data = pd.read_csv('IRIS.csv')
data.head(10)
```

```python
#Выделение тренировочной и тестовой выборки
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, LogisticRegression
X = data[['sepal_length','sepal_width','petal_length','petal_width']]
y = data['species']
X_train, y_train, X_valid, y_valid = train_test_split(X,y,test_size=0.5,random_state=17)
```

```python
#LabelEncoder
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
columnsToEncode = list(data.select_dtypes(include=['category','object']))#Столбцы для кодирования

for feature in columnsToEncode:
    data[feature] = lb.fit_transform(data[feature])#fit-transform

X_enc = data[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
Y_enc = data['species']

x_train, x_test, y_train, y_test = train_test_split(X_enc, Y_enc, test_size=0.20)#train-test-split

data
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
classifier = DecisionTreeClassifier()
classifier.fit(x_train,y_train)
y_pred = classifier.predict(x_test)
print(accuracy_score(y_test,y_pred))
```

```python
#confusion matrix, precision, recall and F1 score
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
